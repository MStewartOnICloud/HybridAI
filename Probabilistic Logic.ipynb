## Probabilistic Logic

In probabilistic logic, we can assign probabilities to logical statements. For example, we can say that the probability of "It will rain tomorrow" is 0.7, which means that there is a 70% chance that it will rain tomorrow. We can also use probabilistic logic to reason about uncertain information. For example, if we know that "If it rains, then the ground will be wet" and "The probability of rain is 0.7", we can infer that "The probability of the ground being wet is 0.7". Probabilistic logic allows us to make informed decisions based on uncertain information and to reason about the likelihood of different outcomes. It is widely used in fields such as artificial intelligence, machine learning, and decision theory to model and reason about uncertainty in various domains.

We need to be able to combine the probab;lities of statements when used in logical operations. For example, if we have two statements A and B, and we know the probabilities of A and B, we can calculate the probability of $A \land B$, $A \lor B$, and $ \neg A$.

### NOT

Given the probability of the statement A is:

$$
p_A
$$

 and we know that either A ot Not A is true, then the probability of $ \neg A$ is:

$$
p_{\neg A} = 1 - p_A \label{eq:neg}
$$

### AND

For two independent statements A and B, the probability of $A \land B$ is:

$$
p_{A \land B} = p_A \cdot p_B \label{eq:and}
$$

this is the same as the probability of two independent events happening at the same time.

### OR

The probability of statement a or statement B is true can be derived from De Morgan's laws:

$$
A \lor B = \neg( \neg A \land \neg B) \label{eq:or}
$$

From (\ref{eg:neg}):
$$
\begin{align}
\begin{split}
p_{\neg A} ={}& 1 - p_A \\
p_{\neg B} ={}& 1 - p_B \\
\end{split}
\end{align}
$$
From (\ref{eq:and}):
$$
\begin{align}
\begin{split}
p_{A \land B} ={}& p_A \cdot p_B \\
\\
p_{\neg A \land \neg B} ={}& p_{\neg A} \cdot p_{\neg B} \\
\end{split}
\end{align}
$$

Rearranging the above, we get:
$$
\begin{align}
\begin{split}
p_{A \lor B} ={}& 1 - ((1 - p_{A}) \cdot (1 - p_{B})) \\
             ={}& 1 - (1 -(p_a + p_b) + p_a \cdot p_b)
\end{split}
\end{align}
$$
